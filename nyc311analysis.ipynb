**.loc[]** vs **.iloc[]** (Label-based selection vs. Position-based selection)

```python
df
```

```python
#.loc[] allows the access of data using labels such as rows and columns
```

```python
# We can access the value in row with label # and column_name using loc[]
df.loc[3, 'complaint_type']
```

```python
# We can also get values for mutiple rows and mutiple columns
df.loc[1:3, ['agency_name', 'complaint_type']]
```

```python
# We can also access the value based on index using iloc[]
#iloc[row #, col #]
df.iloc[0,3]
```

```python
# Access rows 1-2 and columns 0-5, iloc excludes the last value whereas loc includes the last value
df.iloc[1:3, 0:6]
```

```python
#.groupby() allows you to aggregate data based on a column and apply functions like .count(), .sum() or .mean() to name a few
#syntax: df.groupby("column_to_group")["column_to_aggregate"].agg_function()
complaints_by_bourough = df.groupby("borough")["complaint_type"].count()
complaints_by_bourough
```

```python
#.concat(), combines two or more DataFrames along a specified axis (rows or columns)
#axis = 0 is the defult which stacks rows
#axis = 1 Joins columns side by side
df_p1 = df.iloc[:2]
df_p2 = df.iloc[-2:]
df_combined = pd.concat([df_p1, df_p2], axis = 0)
df_combined
```

```python
#.pivot_table() is used to create a summary table that restructes data
#syntax: df.pivot_table(index="row_category", columns="column_category", values="numeric_column", aggfunc="sum")
pivot = df.pivot_table(index = "borough", columns = "complaint_type", values = "unique_key", aggfunc = "count", fill_value = 0)
#aggfunc ="count", counts the number of complaints per category
pivot
#this is good for counting complaints by borough and complaint type
```

**Merging Data with different joins**
If you're familiar with SQL and Left Join, Right Join, Inner Join, and Outer Join, .merge() is similar as it works by combining two DataFrames based on a common column

```python
df2 = df[["borough", "complaint_type","unique_key"]]
df2_grouped = df2.groupby("borough")["unique_key"].count().reset_index()
df2_grouped.rename(columns = {"unique_key": "complaint_count"}, inplace = True)

#hypothetical fake population dataset
borough_population = pd.DataFrame({
    "borough": ["BRONX", "BROOKLYN", "MANHATTAN", "QUEENS", "STATEN ISLAND", "NULL"],
    "population": [1472654, 2648771, 1628706, 2333054, 495747, 0]
})

print(df2_grouped)
print(" ")
print(borough_population)
```

**Left Join (how = "left)**: Keeps all rows from the left DF and only matching rows from the right
**Also note that any missing values are filled with NaN

```python
left_merge = df2_grouped.merge(borough_population, on = "borough", how = "left")
left_merge
```

**Right Join (how = "right")**: Keeps all rows from the right DF and onlt matching rows from the left

```python
right_merge = df2_grouped.merge(borough_population, on = "borough", how = "right")
right_merge
```

**Inner Join (how = "inner")**: Keeps only matching rows from both DFs

```python
inner_merge = df2_grouped.merge(borough_population, on = "borough", how = "inner")
inner_merge
```

**Outer Join (how = "outer")**: Keeps all rows from both DFs

```python
outer_merge = df2_grouped.merge(borough_population, on = "borough", how = "outer")
outer_merge
```

**Cleaning data**

Sometimes there are some values being missing and we would call those values Null values

```python
#.dropna(), similarly to .drop(), drops rows or columns WITH missing values
```

```python
# Check the total number of null values
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# We can clean the data by using dropna()
newdf1 = df.dropna()
null_counts = newdf1.isnull().sum()
print(null_counts)
```

```python
# Aside from dropping missing values, we can also replace the missing values with a specific value
# Let's check out the closed_date column, here we can see a few NaNs
df['closed_date']
```

```python
#.fillna() helps with filling in missing values using specified values or methods
```

```python
# Now, with fillna(), we can replace the NaN with a specific value
df['closed_date'].fillna("Open")
```
